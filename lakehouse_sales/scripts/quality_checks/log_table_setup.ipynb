```python
#create a schema sales.audit
spark.sql("""
          create schema if not exists sales.audit
          """)
```

```python
#create delta table
spark.sql("""       
            CREATE TABLE IF NOT EXISTS sales.audit.quality_log (
                run_id          STRING,
                run_timestamp   TIMESTAMP,
                layer           STRING,          -- bronze / silver / gold
                table_name      STRING,
                check_name      STRING,
                check_type      STRING,          -- completeness, validity, consistency, etc.
                status          STRING,          -- PASSED / FAILED / WARNING
                severity        STRING,          -- CRITICAL / HIGH / MEDIUM / LOW
                records_total   LONG,
                records_failed  LONG,
                failure_rate    DOUBLE,
                details         STRING,          -- JSON with extra context
                pipeline_run_id STRING           -- link to your job run
            )
            USING DELTA -- ensure creating delta table not just parquet table
            PARTITIONED BY (layer) -- partition by layer to make queries filtered by layer faster
        """
)
```
